```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset with the correct delimiter
file_path = "student-mat.csv"
df = pd.read_csv(file_path, delimiter=";")

# Display the first few rows
display(df.head())

# Data Exploration
print("\nMissing values per column:\n", df.isnull().sum())
print("\nColumn Data Types:\n", df.dtypes)
print("\nDataset Shape:", df.shape)

# Data Cleaning
# Handling missing values (if any)
df.fillna(df.median(numeric_only=True), inplace=True)
# Removing duplicate entries
df.drop_duplicates(inplace=True)

# Data Analysis
# 1. Average final grade (G3)
avg_g3 = df['G3'].mean()
print("\nAverage Final Grade (G3):", avg_g3)

# 2. Students scoring above 15 in G3
above_15_count = (df['G3'] > 15).sum()
print("\nNumber of students scoring above 15 in G3:", above_15_count)

# 3. Correlation between study time and final grade
correlation = df['studytime'].corr(df['G3'])
print("\nCorrelation between study time and G3:", correlation)

# 4. Gender-based average final grade
avg_g3_gender = df.groupby('sex')['G3'].mean()
print("\nAverage Final Grade by Gender:\n", avg_g3_gender)

# Data Visualization
plt.figure(figsize=(12, 4))

# 1. Histogram of final grades
plt.subplot(1, 3, 1)
sns.histplot(df['G3'], bins=10, kde=True, color='blue')
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("G3 Score")
plt.ylabel("Count")

# 2. Scatter plot of study time vs final grade
plt.subplot(1, 3, 2)
sns.scatterplot(x=df['studytime'], y=df['G3'], color='red')
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time (Hours)")
plt.ylabel("G3 Score")

# 3. Bar chart for average scores by gender
plt.subplot(1, 3, 3)
sns.barplot(x=avg_g3_gender.index, y=avg_g3_gender.values, palette='viridis')
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average G3 Score")

plt.tight_layout()
plt.show()
```
